```python
import matplotlib.pyplot as plt

def ak(k):
    return k**2 - k + 1

def bk(k):
    if k % 2 == 0:
        return k**2 + 1
    else:
        return k**2

def ck(k):
    if k % 2 == 0:
        return k**2
    else:
        return k**2 + 1

k_values = list(range(1, 101))

ak_values = [ak(k) for k in k_values]
bk_values = [bk(k) for k in k_values]
ck_values = [ck(k) for k in k_values]

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(k_values, ak_values, marker='o', linestyle='-', color='b')
plt.title('a_k')
plt.xlabel('k')
plt.ylabel('a_k')

plt.subplot(1, 3, 2)
plt.plot(k_values, bk_values, marker='o', linestyle='-', color='r')
plt.title('b_k')
plt.xlabel('k')
plt.ylabel('b_k')

plt.subplot(1, 3, 3)
plt.plot(k_values, ck_values, marker='o', linestyle='-', color='g')
plt.title('c_k')
plt.xlabel('k')
plt.ylabel('c_k')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

def Sa(k):
    return (k**3 + 2*k) / 3

def Sb_odd(k):
    return (k*(k+1)*(2*k+1)) / 6 + (k-1) / 2

def Sc_even(k):
    return (k*(k+1)*(2*k+1)) / 6 + k / 2

k_values = list(range(1, 101))
k_odd = list(range(1, 101, 2))
k_even = list(range(2, 101, 2)) 

Sa_values = [Sa(k) for k in k_values]
Sb_values = [Sb_odd(k) for k in k_odd]  
Sc_values = [Sc_even(k) for k in k_even]  

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(k_values, Sa_values, marker='o', linestyle='-', color='b')
plt.title('Sa(k)')
plt.xlabel('k')
plt.ylabel('Sa(k)')

plt.subplot(1, 3, 2)
plt.plot(k_odd, Sb_values, marker='o', linestyle='-', color='r')
plt.title('Sb(k)')
plt.xlabel('k')
plt.ylabel('Sb(k)')

plt.subplot(1, 3, 3)
plt.plot(k_even, Sc_values, marker='o', linestyle='-', color='g')
plt.title('Sc(k)')
plt.xlabel('k')
plt.ylabel('Sc(k)')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

def S(k):
    if k % 2 == 1:
        return (k**3 + 2*k) / 3 + (k*(k+1)*(2*k+1)) / 6 + (k-1) / 2 - 1
    else:
        return (k**3 + 2*k) / 3 + (k*(k+1)*(2*k+1)) / 6 + k / 2

k_values = list(range(1, 101))

S_values = [S(k) for k in k_values]

plt.figure(figsize=(8, 6))
plt.plot(k_values, S_values, marker='o', linestyle='-', color='purple')
plt.title('S(k)')
plt.xlabel('k')
plt.ylabel('S(k)')
plt.grid(True)
plt.show()
```
