```python
import numpy as np

def simulate_game(num_trials):
    count = 0
    for _ in range(num_trials):
        tosses = np.random.randint(1, 3, size=24)  
        cumulative_sum = np.cumsum(tosses)
        if np.any(cumulative_sum == 24): 
            count += 1
    probability = count / num_trials
    return probability

num_trials = 100000000 

probability = simulate_game(num_trials)
print("Probability of obtaining a sum of 24 at any point within 24 tosses at least once in", num_trials, "trials:", probability)
```

```python
import matplotlib.pyplot as plt
from scipy.special import comb

n_values = range(0, 25, 2)

probabilities = [comb(12 + n // 2, n) / (2 ** (12 + n // 2)) for n in n_values]

plt.plot(n_values, probabilities, marker='o')
plt.xlabel('Number of 1s (n)')
plt.ylabel('Probability')
plt.title('Probability of Particular Combination vs. Number of 1s')
plt.grid(True)
plt.show()
```
